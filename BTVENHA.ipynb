```python
import turtle
pen = turtle.Turtle()
turtle.Screen().bgcolor("Lightblue")

# ve nha
pen.fillcolor("yellow")
pen.left(90)
pen.begin_fill()
pen.backward(150)
pen.left(120)
pen.forward(100)
pen.right(120)
pen.forward(150)
pen.right(60)
pen.forward(100)
pen.backward(100)
pen.end_fill()

pen.fillcolor("blue")
pen.right(30)
pen.begin_fill()
pen.backward(120)
pen.right(90)
pen.forward(150)
pen.left(90)
pen.forward(120)
pen.left(90)
pen.forward(150)
pen.end_fill()

pen.fillcolor("pink")
pen.left(45)
pen.begin_fill()
pen.forward(85)
pen.left(90)
pen.forward(85)
pen.left(135)
pen.forward(120)
pen.left(135)
pen.forward(85)
pen.end_fill()

pen.fillcolor("orange")
pen.right(115)
pen.begin_fill()
pen.forward(90)
pen.right(53)
pen.forward(75)
pen.right(117)
pen.forward(100)
pen.end_fill()

# cua cho nha
pen.penup()
pen.goto(-30,-70)
pen.pendown()
pen.fillcolor("brown")
pen.begin_fill()
pen.forward(30)
pen.left(60)
pen.forward(30)
pen.left(120)
pen.forward(30)
pen.left(60)
pen.forward(30)
pen.end_fill()

# ve cua chinh
pen.penup()
pen.goto(-170,-100)
pen.pendown()
pen.fillcolor("green")
pen.begin_fill()
pen.backward(100)
pen.right(90)
pen.forward(40)
pen.left(90)
pen.forward(100)
pen.left(90)
pen.forward(40)
pen.end_fill()

#ve soi chieu sang
pen.penup()
pen.goto(200,150)
pen.pendown()
pen.right(90)
pen.forward(50)
pen.backward(75)
for i in range (4):
    pen.left(45)
    pen.forward(60)
    pen.backward(60)
    pen.left(45)
    pen.forward(75)
    pen.backward(75)

# ve ong mat troi
pen.forward(25)
pen.left(90)
pen.fillcolor("yellow")
pen.begin_fill()
pen.circle(30)
pen.end_fill()

# ve cay
pen.penup()
pen.goto(250,0)
pen.pendown()
pen.fillcolor("green")
pen.begin_fill()
for i in range (2):
    pen.pensize(1)
    pen.left(45)
    pen.forward(85)
    pen.pensize(3)
    pen.left(135)
    pen.forward(120)
    pen.left(135)
    pen.forward(85)
    pen.left(45)
    pen.penup()
    pen.goto(250,-60)
    pen.pendown()
pen.penup()
pen.goto(250,-120)
pen.pendown()
pen.pensize(1)
pen.left(45)
pen.forward(85)
pen.pensize(3)
pen.left(135)
pen.forward(120)
pen.left(135)
pen.forward(85)

pen.end_fill()

# ve than cay
pen.penup()
pen.goto(250,-240)
pen.pendown()
pen.fillcolor("brown")
pen.begin_fill()
pen.pensize(1)
pen.left(45)
pen.backward(10)
pen.forward(20)
pen.right(90)
pen.forward(60)
pen.right(90)
pen.forward(20)
pen.right(90)
pen.forward(60)

pen.end_fill()

turtle.done()
```
